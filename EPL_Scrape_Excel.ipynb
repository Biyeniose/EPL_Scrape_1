```python
driver.find_element(By.XPATH, "(//a[contains(text(),\'Manchester City\')])[2]").click()
driver.find_element(By.LINK_TEXT, "Detailed squad").click()
```

```python
table = driver.find_element(By.CLASS_NAME,'items')
table_1 = table.find_element(By.TAG_NAME,'tbody')
for row in table_1.find_elements(By.TAG_NAME,'tr'):
    print("------------------")
    for name in row.find_elements(By.CLASS_NAME, 'posrela'):
        for name_1 in name.find_elements(By.CLASS_NAME, 'hauptlink'):
                name_2 = name_1.find_element(By.TAG_NAME,'a')
                print(name_2.text)
                
                for a_1 in row.find_elements(By.CLASS_NAME, 'zentriert'):
                    for a_2 in a_1.find_elements(By.TAG_NAME,'img'):
                            print(a_2.get_attribute('alt'))
                           
```
